```python
with open("the-verdict.txt", "r", encoding="utf-8") as file:
    text_ver = file.read()

print("Total characters:", len(text_ver))
print(text_ver[:99])
```

```python
# we use python's regular expression library for splitting text

import re

text = "Hello world, yoooo."
result = re.split(r'(\s)', text)

print(result)
```

```python
# for removing whitespaces

result = [item for item in result if item.strip()]

print(result)
```

Removing whitespaces reduces the memory and computing
requirements. However, keeping whitespaces can be useful if we train models that
are sensitive to the exact structure of the text (for example, Python code, which is
sensitive to indentation and spacing). Here, we remove whitespaces for simplicity
and brevity of the tokenized outputs.

```python
text = "Hello, world. Is this-- a test?"
result = re.split(r'([,.:;?_!"()\']|--|\s)', text)
result = [item for item in result if item.strip()]
print(result)
```

```python
# for the verdict

preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text_ver)
preprocessed = [item for item in preprocessed if item.strip()]
print(preprocessed[:30])
```
